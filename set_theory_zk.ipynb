## Cartesian Product

> We can define a set such that every element from one set is one part of an ordered pair with an element from another set

- i.e.
    - $A = \{{1,2,3}\}$
    - $B = \{{x,y,z}\}$
    - $A \times B = \{{(1,x),(1,y),(1,z),(2,x),(2,y),(2,z),(3,x),(3,y),(3,z)}\}$

```python
# This code displays the table from the example above

import pandas as pd
from itertools import product

# Example sets
setA = {1, 2, 3}
setB = {'x', 'y', 'z'}

# Calculate the Cartesian product
cartesian_product = product(setA, setB)

# Create a DataFrame with setB elements as rows and setA elements as columns
df = pd.DataFrame(cartesian_product, columns=['SetA', 'SetB'])
df['Ordered Pair'] = df.apply(lambda row: f'({row["SetA"]}, {row["SetB"]})', axis=1)
pivot_table = df.pivot(index='SetB', columns='SetA', values='Ordered Pair')
pivot_table = pivot_table.rename_axis(index=None)
pivot_table.columns = list(setA)

# Display the DataFrame
pivot_table
```

### Exercise 7:

Define a mapping (function) from integers $n \in 1,2,3,4,5,6$ to the set {even, odd}.

### Answer 7:

```python
# Exercise 7

n = np.arange(1,7,1)

def mapping_even_odd(nums):
    setEven = []
    setOdd = []
    setFinal = set()
    for num in nums:
        if num % 2 == 0:
            setEven += [num]
        else:
            setOdd += [num]
    setFinal.add(str(setOdd))
    setFinal.add(str(setEven))
    return setFinal

result = mapping_even_odd(n)

print('Set: ', result)
print('Even Set: ', list(result)[1])
print('Odd Set: ', list(result)[0])
```

## Injective, Surjective, and Bijective functions

### Definition: Injective Function

An injective function means that the elements in the codomain have at **most** one preimage.

This means that every element from the codomain gets hit at most once if it gets hit.

#### Example: Injective Function

Let $f: \bar X \to \bar Y, X = domain, Y = codomain$ 

\begin{matrix}
\bar X & \bar Y \\
1  & \to 3\\
2  & \to 4\\
   &     5
\end{matrix}


### Definition: Surjective Function

A surjective function means the elements in the codomain has at **least** one preimage.

This means that every element from the codomain gets hit at least once if it gets hit.

#### Example: Surjective Function

Let $f: \bar X \to \bar Y, X = domain, Y = codomain$ 

\begin{matrix}
\bar X & \bar Y \\
1  & \to 4\\
   & \nearrow \\
2  & \\
\\
3  & \to 5
\end{matrix}

### Definition: Bijective Function

A function is bijective if and only if it is injective and surjective

This means that everyelemtn in the codomain gets hit at **least** one and that everything gets hit at **most** once

#### Example: Bijective Function

Let $f: \bar X \to \bar Y, X = domain, Y = codomain$ 

\begin{matrix}
\bar X & \bar Y \\
1  & \to 4\\
2  & \to 5\\
3  & \to 6
\end{matrix}

## A Cartesian Product of a set with itself

It should be no surprise that instead of doing a cartesian product between $A$ and $B$, you can do a cartesian product between $A$ and $A$. This is just mapping a set to itself.

This is actually a more abstract form of what we traditionally think of as functions between integers.


For example, $y = x^2$ (over positive integers) can be visualized in set theoretic terms as the subset of $\mathbb{Z}+ \, \times    \,\, \mathbb{Z}+$

```python
setA = np.arange(1,11,1)

cartesian_product = product(setA, setA)

df = pd.DataFrame(cartesian_product, columns=['SetVertical', 'SetHorizontal'])

df['Ordered Pair'] = df.apply(lambda row: (row["SetVertical"], row["SetHorizontal"]), axis=1)

pivot_table = df.pivot(index='SetHorizontal', columns='SetVertical', values='Ordered Pair')
pivot_table = pivot_table.rename_axis(index=None)
pivot_table.columns = list(setA)

# Define a custom function for styling

bold_pairs = [(i,i*i) for i in range(11)]
    
def bold_ordered_pairs(pair):
    return f'font-weight: bold' if pair in bold_pairs else ''

# Apply the styling to the entire DataFrame
styled_table = pivot_table.style.applymap(bold_ordered_pairs)

# # Display the styled DataFrame
styled_table
```

## Set Relations

### Definition: Relation

Taking a subset of a cartesian product is called a **relation**

For example,

If we have to sets A and B and we take a subset of the cartesian product, A X B.

We say than that an element from A is related to an element from B if they have an ordered pair


## A binary operator in set theoretic terms

A binary operator is a function from $A \times A \to A$. Basically, we take every possible pair from $A$ (the cartesian product of $A$ with itself) and map it to $A$.


Letâ€™s use an example of the set $\{0,1,2\}$ with binary operator addition modulo 3. First we take the setâ€™s cartesian product with itself:

```python
A = np.arange(0,3,1)

cartesian_product = product(A, A)

df = pd.DataFrame(cartesian_product)

df["Ordered Pair"] = df.apply(lambda row: (row[1], row[0]), axis=1)

pivot_table = df.pivot(index=0, columns=1, values='Ordered Pair')
pivot_table = pivot_table.rename_axis(index=None)
pivot_table.columns = list(A)

df_ordered_pair = df["Ordered Pair"]
pivot_table
```

Take the cartesian product of this new set of pairs with the original set

```python
ordered_pairs = [ i for i in df_ordered_pair]

cartesian_product = product(A, ordered_pairs)

df = pd.DataFrame(cartesian_product)

df["Ordered Pair"] = df.apply(lambda row: (row[1], row[0]), axis=1)

pivot_table = df.pivot(index=0, columns=1, values='Ordered Pair')
pivot_table = pivot_table.rename_axis(index=None)
pivot_table.columns = ordered_pairs

pivot_table.T
```

Take the subset of that which defines our binary operator addition modulo 3

```python
def binary_op_add_mod3(df):
    
    # Define a custom function for styling
    bold_pairs = []
    
    for i in range(len(df)):
        if sum(df[1][i])%3 == df[0][i]:
            bold_pairs += [(df[1][i],df[0][i])]
            
    return bold_pairs

bold_pairs = binary_op_add_mod3(df)

def bold_ordered_pairs(pair):
    return f'font-weight: bold' if pair in bold_pairs else ''

# Apply the styling to the entire DataFrame
styled_table = pivot_table.style.applymap(bold_ordered_pairs)

# # Display the styled DataFrame
styled_table
```

## Constructing valid binary operations

When it comes to binary operators, we are not allowed to take a subset of $A \times A$ before mapping that to $A$. Binary operators must accept all members of group A as its inputs. We of course must take a subset of the ordered pairs between $A \times A$ and $A$ because each pair from $A \times A$ must map to exactly one $A$.
